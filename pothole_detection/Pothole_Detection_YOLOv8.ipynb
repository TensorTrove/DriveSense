```python
import zipfile
import os

dataset_path = 'Pothole.v1-raw.yolov8.zip'
extract_path = 'pothole_dataset'

with zipfile.ZipFile(dataset_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset extracted to:", extract_path)
```

```python
from ultralytics import YOLO
import os

data_path = extract_path
model = YOLO('yolov8n.pt')

model.train(data=os.path.join(data_path, 'data.yaml'), epochs=50, imgsz=640)

model.save('pothole_detector.pt')
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('pothole_detector.pt')

input_video_path = 'pothole.mp4'
output_video_path = 'pothole_annotated.mp4'

cap = cv2.VideoCapture(input_video_path)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    print(results)

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            conf = box.conf[0]
            cls = int(box.cls[0])
            label = f"{model.names[cls]} {conf:.2f}"

            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    out.write(frame)

cap.release()
out.release()

print("Annotated video saved to:", output_video_path)
```
